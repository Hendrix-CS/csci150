Today, we will learn about a new data type, `boolean`. At heart, a boolean is simply a statement which is either `True` or `False.`

The word *boolean* comes from George Boole, a 19th Century English mathematician and logician. Among other accomplishments was his formalization of the algebra of logical statements -- that is, given a complicated statement (or set of statements) where we know the truth value of each individual part, what can we say about the whole:

Suppose that we know:

* "Today is Friday" is `True`
* "It is currently raining" is `False`

Then, what can be said about the statement "Today is Friday and it is currently raining" ?

What about "Today is Friday or it is currently raining" ?

We can combine various boolean statements using the special three operators `and`, `or`, and `not`.

Let's start with `not`:

| $P$ | `not` $P$ |
|:-----|:-----------|
|`True`|`False`|
|`False`|`True`|

```python
not 5 == 7
```

`and`:

|$P$ | $Q$ | $P$ `and` $Q$ |
|:---|:---|:---|
|`True`|`True`|`True`|
|`True`|`False`|`False`|
|`False`|`True`|`False`|
|`False`|`False`|`False`|


`or`:

|$P$ | $Q$ | $P$ `or` $Q$ |
|:---|:---|:---|
|`True`|`True`|`True`|
|`True`|`False`|`True`|
|`False`|`True`|`True`|
|`False`|`False`|`False`|

```python
( 5 == 7) and ('a' == 'a')
```

Like `+` and `*`, there is an implied precedence order -- in fact, we can think of `or` as `+` and `and` as `*`, so that:

`True or True and False` is interpretted as `True or (True and False)`, just like $2 + 3 \times 5 = 2 + (3 \times 5)$.

You should *not* trust this, however -- just put parentheses!

```python
(True and (not (False or False))) and (False or (True or False))
```
