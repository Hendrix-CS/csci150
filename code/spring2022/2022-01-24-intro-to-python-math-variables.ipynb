Example: The Ideal Gas law

In Chemistry, you often learn the Ideal Gas Law, which states $$PV = nRT,$$ where $P$ is the pressure, meaured in kilo pascals, $V$ the volume in liters, $n$ is the number of moles of the gas, $R =8.314$ is the gas constant, and $T$ the temperature measured in Kelvin.

Determine the pressure of a 2 moles of a gas with a volume of 2.50 liters at 300 K.

Do some algebra:

```python
V = 3.50
n = 2
R = 8.314
T = 300
```

```python
P = (n * R * T) / V
```
