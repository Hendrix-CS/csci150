Hailstone problem:

Consider the following: $f(x) = \begin{cases} \frac{x}{2},& \text{ if } x \text{ is even} \\ 3x + 1,& \text{ if } x \text{ is odd}\end{cases}$

We will call this function the "hailstone" function. Let's try a few starting numbers and see what happens

8: 8, 4, 2, 1, then back to 4, 2, 1, back to 4, ...

10: 10, 5, 16, 8, 4, 2, 1, ...

7: 7, 22, 11, 34, 17, 52, 26, 13, 40, 20, 10, 5, 16, 8, 4, 2, 1

It is called "hailstone" since numbers generally fall toward 1, but occasionally, rise up, somewhat analogous to a "real" hailstone.

Does every given starting positive integer eventually reach 1?

Collatz Conjecture -- claim is that each number will eventually reach 1

Can a computer solve this problem? aren't there *infinitly* many?

As of 2020 (according to Wikipedia, which is usually reliable about such details), all starting numbers up to $2^{68} \approx 2.95 \times 10^{20}$ have been checked, and all do eventually reach 1.
