As review, write a function `fizzbin` which takes in three parameters:
* `s`, a string
* `hour` an integer, and
* `is_am`, a boolean

Here are the rules for fizzbin. If the time is between 3am and 1pm (inclusive) *and* your string is smaller than `"j"`, you win! You also win if instead the time is between 6pm and 8pm (also inclusive), and your string is between `"n"` and `"qzzzz"` (inclusive).

Your function should return `True` if you win and `False` otherwise. You can assume that the value of `hour` will always be an integer between 1 and 12.





```python
fizzbin('a',6,True)
```

My solution is written below.

You should try this on you own and then we will go over it.

```python
def fizzbin(s: str, hour: int, is_am: bool) -> bool:
    if (hour >= 3 and is_am) or ((hour == 1 or hour == 12) and not is_am):
        if s < 'j':
            return True
        else:
            return False
        
    elif 6 <= hour <= 8 and not is_am:
        if 'n' <= s <= 'qzzzz':
            return True
        else:
            return False
        
    else:
        return False
```
